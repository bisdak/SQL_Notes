```python
import sqlite3
import pandas as pd
from pandas import DataFrame
```

```python
def get_data(query: str, db: str = "northwind.db") -> DataFrame:
    # Connect to database
    with sqlite3.connect(db) as conn:
        # Execute the query and store the result in a DataFrame
        df = pd.read_sql_query(query, conn)
    return df
```

```python
df = get_data("SELECT * FROM Employees")
df
```

```python
tables = get_data("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name;")
tables
```

```python
categories_regions = get_data("SELECT * FROM Categories, Regions")
df
```

```python
# DISTINCT rows
distinct_rows = get_data("SELECT DISTINCT City FROM Customers")
distinct_rows
```

```python
greater_than_10 = get_data(
    "SELECT ProductName, UnitPrice FROM Products WHERE UnitPrice > 10 ORDER BY UnitPrice DESC"
)
greater_than_10
```

```python
data = get_data("SELECT * FROM Orders WHERE OrderDate > '2017-05-01'")
data
```

```python
# Ranges
between_5_and_10 = get_data("SELECT * FROM Products WHERE UnitPrice BETWEEN 5 and 10")
between_5_and_10
```

```python
# INNER JOIN
data = get_data(
    """
    SELECT Products.ProductName, Categories.CategoryName
    FROM Categories
    INNER JOIN Products ON Categories.CategoryID = Products.CategoryID
    WHERE Categories.CategoryName = 'Beverages'
    """
)
data
```

```python
# Alternative to INNER JOIN or JOIN
data = get_data(
    """
    SELECT Products.ProductName, Categories.CategoryName
    FROM Categories, Products 
    WHERE Categories.CategoryID=Products.CategoryID AND Categories.CategoryName = 'Beverages'
    """
)
data
```
