```python
import sqlite3
import pandas as pd
from pandas import DataFrame
```

```python
DB = "northwind.db"
```

```python
def get_data(query: str, db: str) -> DataFrame:
    # Connect to database
    with sqlite3.connect(db) as conn:
        # Execute the query and store the result in a DataFrame
        df = pd.read_sql_query(query, conn)
    return df
```

```python
df = get_data("SELECT * FROM Employees", DB)
df
```

```python
tables = get_data("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name;", DB)
tables
```
